```python
dataset = pd.read_csv('movies.csv')
db = GraphDatabase.driver(uri='bolt://localhost:7687', auth=('neo4j', '1234'))
session = db.session()
```

```python
dataset.head()
```

```python
genres_list = []
for movie in dataset.iloc:
    genres_list = movie[2].split("|")
    for i in genres_list:
        session.run("call gds.graph.create.cypher('gdb', 'MERGE (n:Movie {id: $id, name: $movie})', 'MERGE (m:Genre {name: $genre}) CREATE (n) -[:TO {dist: 5}] -> (m)')", id=str(movie[0]), movie=movie[1], genre=i)
```

```python
df=pd.read_csv('ratings.csv')
df.head()
```

```python
for a in df['userId'].unique():
  session.run("CREATE (n:User {id:$id})",id=int(a))
```

```python
for row in df.loc:
    session.run("MATCH (n:User {id: $id}), (m:Movie {id: $movieId}) CREATE (n) -[:VOTE {dist: 5, rating: $rating,ts: $ts}] -> (m)", id=int(row[0]), movieId=str(int(row[1])),rating=str(row[2]), ts=str(row[3]))
```

```python
def createNode(nodeName,propName):
    query = "CREATE (n: "+ nodeName  + " { name: $name }) RETURN n"
    session.run(query, name=propName)
createNode("A","B")
```

```python
def deleteNodeAndRel(nodeName):
    query = "MATCH (n:"+nodeName+") DETACH DELETE n "
    session.run(query,name=nodeName)
deleteNodeAndRel("A")
```

```python
def nodeInfo(nodeName):
    query=("MATCH (n:"+nodeName+") RETURN n ")
    return session.run(query)
movie=nodeInfo("Movie")
# node propertylerine erimek
for i in movie:
    print("id: "+i[0]['id'],"name: "+i[0]['name'])
```

```python
def nodeCount(nodeName):
    query=("MATCH (n:"+nodeName+") RETURN COUNT(n) ")
    return session.run(query)
movie=nodeInfo("Movie")
for i in movie:
    print(i[0])
```
